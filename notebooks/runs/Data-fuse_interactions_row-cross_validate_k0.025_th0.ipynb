```python
# importing modules
import matplotlib.pyplot as plt
import numpy as np
from skfusion import fusion
from cross_validation import *
import pickle
from math import ceil
import time

from scipy.stats import ranksums
%matplotlib inline
plt.rcParams['figure.figsize'] = (15.0, 8.0)
plt.style.use('ggplot')
```

# Matrices Import

```python
prot_rna = pickle.load(open('../matrices/pickles/cr_z-score_raw.pkl', 'rb'))
secretion_prot = pickle.load(open('../matrices/pickles/signalP.pkl', 'rb'))
rna_expression = pickle.load(open('../matrices/pickles/MA_exp.pkl', 'rb'))
s4te_prot = np.load(open('../matrices/pickles/S4TE.npy', 'rb'))
OG_prot = np.load(open('../matrices/pickles/OG_prot.npy', 'rb')) # this is orthology function
rna_GO = np.load(open('../matrices/pickles/HSgene_GO.npy', 'rb'))

gproOG_prot = np.load(open('../matrices/pickles/aaprot_gprobOG.npy', 'rb'))
hsgene_gproOG = np.load(open('../matrices/pickles/HSgene_GO.npy', 'rb'))
aaexp_prot = np.load(open('../matrices/pickles/aa_fc_expressions.npy', 'rb'))
```

# Data transformation

```python
# remove data for genes with no catrapid prediction
selection = prot_rna.any(axis=1)
prot_rna = prot_rna[selection,:]
secretion_prot = secretion_prot[:, selection]
s4te_prot = s4te_prot[:, selection]
OG_prot = OG_prot[:, selection]
gproOG_prot = gproOG_prot.T[:, selection]
# Addd pseudo rows 
# TODO: remove when fixed!
secretion_prot = np.vstack((secretion_prot, secretion_prot, secretion_prot, secretion_prot, secretion_prot))

rna_expression = rna_expression.reshape(rna_expression.shape[0],1)
rna_expression = np.hstack((rna_expression, rna_expression, rna_expression, rna_expression, rna_expression))

aaexp_prot = np.reshape(aaexp_prot, (1, aaexp_prot.size))
aaexp_prot = np.vstack((aaexp_prot, aaexp_prot, aaexp_prot, aaexp_prot, aaexp_prot))[:, selection]

# Add sum to S4te Results
s4te_prot[1,:] = s4te_prot[1,:] * 3
s4te_sum = s4te_prot.sum(axis=0) / s4te_prot.sum(axis=0).max()
s4te_prot = np.vstack((s4te_prot,s4te_sum))

# Display shapes
print('secretion_prot:', secretion_prot.shape)
print('rna_expression:', rna_expression.shape)
print('prot_rna:', prot_rna.shape)
print('s4te_prot:', s4te_prot.shape)
print('OG_prot:', OG_prot.shape)
print('gproOG_prot', gproOG_prot.shape)
print('hsgene_gproOG', gproOG_prot.shape)
print('aaexp_prot', aaexp_prot.shape)
print('rna_GO', rna_GO.shape)
```

# Data fusion setup

```python
# SETUP of variables!
k_perc = 0.025
auc_th = 0

# log file
log_file = open('./logs/CR_reconstruct_auc{}_kperc{}.txt'.format(auc_th, k_perc), 'a')


print(('#############'
      'New analysis'
      '#############'
      '\nran on: {}' 
      '\n######################################'
      '\nAUC_th = {}'
      '\nK_perc = {}'
      '\n######################################').format(time.strftime("%d.%m.%Y %H:%M:%S"),auc_th, k_perc), file=log_file)
```

```python
k_select = lambda x: max((ceil(x * k_perc), 2))

# Define names
secretion = fusion.ObjectType('SignalP', k_select(secretion_prot.shape[0]))
prot = fusion.ObjectType('AAprot', k_select(prot_rna.shape[0]))
genes = fusion.ObjectType('HSgenes', k_select(prot_rna.shape[1]))
GO = fusion.ObjectType('GOterms', k_select(rna_GO.shape[1]))

# TODO: lower when possible
expr = fusion.ObjectType('HSexpression', k_select(rna_expression.shape[1]))
t4e = fusion.ObjectType('Type4eff', k_select(s4te_prot.shape[0]))
OGfg = fusion.ObjectType('orthologs_group_functions', k_select(OG_prot.shape[0]))
gproOG = fusion.ObjectType('orthologs_groups', k_select(gproOG_prot.shape[0]))
aaexp = fusion.ObjectType('AAexpression', k_select(aaexp_prot.shape[0]))

# Define relations
relations = [fusion.Relation(secretion_prot, secretion, prot),
             fusion.Relation(prot_rna, prot, genes, name='catrapid'),
             fusion.Relation(rna_expression, genes, expr),
             fusion.Relation(OG_prot, OGfg, prot),
             fusion.Relation(gproOG_prot, gproOG, prot),
             fusion.Relation(hsgene_gproOG, genes, gproOG),
             fusion.Relation(aaexp_prot, aaexp, prot),
             fusion.Relation(rna_GO, genes, GO)]

all_object = [secretion, prot, genes, GO, expr, t4e, OGfg, gproOG, aaexp]
for objx in all_object:
    print(objx, objx.rank)
```

# Validation Functions

```python
def shuffle_rows(my_array, ssed = 100):
    """
    Returns array with rows shuffled
    """
    np.random.seed(ssed)
    sh_idx = np.arange(my_array.shape[0])
    np.random.shuffle(sh_idx)
    return my_array[sh_idx,:]


def shuffle_cols(my_array, ssed = 100):
    """
    Returns array with rows shuffled
    """
    np.random.seed(ssed)
    sh_idx = np.arange(my_array.shape[1])
    np.random.shuffle(sh_idx)
    return my_array[:, sh_idx]

def compare_boxplot(sample1, sample2, s1point, s2point, abline=None):
    """
    Draws boxplots give two samples
    s1point, s2point: selected value to plot
    """
    fig = plt.figure()
    ax = fig.add_subplot(111)
    flierprops = dict(marker='o', markerfacecolor='blue', linestyle='none');
    boxprops= dict( color='blue', linewidth=2)
    capprops = dict(color='blue', linewidth=2)
    medianprops = dict(color='blue', linewidth=2)
    my_bp = plt.boxplot(list(sample1) + [[]] + list(sample2), flierprops=flierprops, boxprops=boxprops, capprops=capprops, patch_artist=True);

    xpoints = list(range(1, len(s1point) + 1)) + list(range(len(s1point) + 2, len(s1point) + len(s1point) + 2))
    plt.plot(xpoints, s1point + s2point,  '*', color='black', markersize=12)
    ax.set_xticklabels(['Podatki; CV {}'.format(i) for i in range(1,11)] + 
                       [''] +
                       ['Permutirani\npodatki; CV {}'.format(i) for i in range(1,11)], rotation='vertical');

    for box in my_bp['boxes'][:10]:
        box.set( facecolor= 'green', alpha = 0.5 )
    for box in my_bp['boxes'][10:]:
        box.set( facecolor= 'blue', alpha = 0.5 )
    
    if abline is not None:
        plt.plot([0, len(s1point) + len (s2point) + 2] , [abline, abline], color='black', alpha = 0.5)
    
    ax.text(0.01,0.01, 'P vrednost: {:.5f} (Wilcoxonov test)'
               .format(ranksums(sample1.flatten(), sample2.flatten())[1]), transform=ax.transAxes)


def row_cross_validate(relations, cv_fold=10, th=2.5, random_state=1, n_reruns=5, shuffle_test=None, abline=None):
    """
    Performs cross valiation on give first matrix of relation and 
    returns AUC scores.
    """
    auc_vals = np.zeros((cv_fold, n_reruns))
    cv_count = 0
    orig_matrix = relations[0].data
    best_AUCs = []
    
    if shuffle_test is not None: 
        print("Doing CV on real data.", file = log_file)
    else: 
        print("Doing CV on Permuted Data.", file = log_file)
    # CV loop
    for cv_matrix, elements in matrix_cv_setup(orig_matrix, n_fold=cv_fold, alternative=0, by_row=True):
        
        relations[0].data = cv_matrix

        # Rerun loop
        best_err = float('inf')
        AUC_point = 0
        for rerun in range(n_reruns):
            reconst_mat, error =  dfmf_n_reconstruct(relations, random_state=rerun)
            cur_auc = auc_value(orig_matrix, reconst_mat, elements, treshold=th, by_row=True)
            auc_vals[cv_count, rerun] = cur_auc

            # Update Best point
            if error < best_err:
                best_err = error
                AUC_point = auc_vals[cv_count, rerun]
        
        best_AUCs.append(AUC_point)       
        cv_count += 1

    relations[0].data = orig_matrix

    
    print('all_aucs:',auc_vals, file = log_file)
    print('AUC at optimum:', best_AUCs, file = log_file)
    
    # test on shuffled Data
    if shuffle_test:
        if  shuffle_test == 'row':
            relations[1].data = shuffle_rows(relations[1].data)

        elif shuffle_test == 'col': 
            relations[1].data = shuffle_cols(relations[1].data)
        auc_vals2, best_AUCs2 = row_cross_validate(relations, cv_fold=cv_fold, th=th, n_reruns=n_reruns)
        compare_boxplot(auc_vals, auc_vals2, best_AUCs, best_AUCs2)
    return auc_vals, best_AUCs


def dfmf_n_reconstruct(relationships, random_state=1):
    """
    Given relationships, reconstructs the matrix
    """
    fusion_graph = fusion.FusionGraph(relations)

    fuser = fusion.Dfmf(random_state=random_state, compute_err = True)
    fuser.fuse(fusion_graph)
    
    error = 0
    for fer in fuser.fusion_graph.relations:
        error += sum([np.absolute(fuser.complete(fer) - fer.data).sum()])
    return fuser.complete(fusion_graph['catrapid']), error

print('Done')
```
